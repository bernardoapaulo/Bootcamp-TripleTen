##### Encontre o número de avaliações e a classificação média para cada livro.

```python
query = """
SELECT 
    books.title, 
    subquery.avg_rating, 
    subquery.review_count
FROM (
    SELECT 
        reviews.book_id AS book_id, 
        COUNT(DISTINCT reviews.review_id) AS review_count, 
        AVG(ratings.rating) AS avg_rating
    FROM reviews
    INNER JOIN ratings ON ratings.book_id = reviews.book_id
    GROUP BY reviews.book_id
    ORDER BY review_count DESC, avg_rating DESC
) AS subquery
INNER JOIN books ON subquery.book_id = books.book_id
ORDER BY review_count DESC, avg_rating DESC
LIMIT 10;
"""

result = pd.read_sql(query, con=engine)
print("Top 10 livros com mais avaliações e melhores classificações:")
print(result)
```

##### Identifique a editora que lançou o maior número de livros com mais de 50 páginas.

```python
query = """
SELECT publishers.publisher, COUNT(books.book_id) AS num_books
FROM books
JOIN publishers ON books.publisher_id = publishers.publisher_id
WHERE books.num_pages > 50
GROUP BY publishers.publisher
ORDER BY num_books DESC
LIMIT 1;
"""
result = pd.read_sql(query, con=engine)
print("Editora que lançou o maior número de livros com mais de 50 páginas:")
print(result)
```

O resultado indica que a editora "Penguin Books" lançou o maior número de livros com mais de 50 páginas, com um total de 42 livros.
- Como o critério de páginas foi de mais de 50, isso exclui publicações muito curtas. O número elevado de livros pode sugerir que "Penguin Books" foca em publicações com maior profundidade de conteúdo, o que pode indicar uma preferência ou especialização por livros que tendem a ser mais substanciais ou detalhados.

#### Identifique o autor com a média mais alta de classificação de livros com pelo menos 50 classificações.

```python
query = """
SELECT 
    authors.author, 
    AVG(subquery2.avg_rating) AS final_avg
FROM (
    SELECT 
        books.title, 
        books.author_id, 
        subquery1.avg_rating
    FROM (
        SELECT 
            book_id, 
            COUNT(rating_id) AS rating_cnt, 
            AVG(rating) AS avg_rating
        FROM ratings
        GROUP BY book_id
        HAVING COUNT(rating_id) > 50
    ) AS subquery1
    INNER JOIN books ON books.book_id = subquery1.book_id
) AS subquery2
INNER JOIN authors ON authors.author_id = subquery2.author_id
GROUP BY authors.author
ORDER BY final_avg DESC
LIMIT 5;
"""

result = pd.read_sql(query, con=engine)
print("Top 5 autores com a média mais alta de classificação (mínimo 50 classificações por livro):")
print(result)
```
